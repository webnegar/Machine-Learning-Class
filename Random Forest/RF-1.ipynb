```python

import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt


from farsi_matplotlib import fa, set_farsi_font
set_farsi_font()  # فعال‌سازی فونت فارسی


# --- داده‌ها ---
data = {
    'اخلاق': [9,8,7,6,5,4,9,7,6,8],
    'ساعت_کار': [50,48,42,40,38,35,55,45,50,46],
    'پروژه': [6,5,4,3,2,1,7,5,2,4],
    'پاداش': ['زیاد','زیاد','متوسط','متوسط','کم','کم','زیاد','متوسط','کم','متوسط']
}
df = pd.DataFrame(data)

# --- آماده‌سازی داده ---
X = df[['اخلاق','ساعت_کار','پروژه']]
y = df['پاداش']

# --- ساخت مدل ---
clf = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
clf.fit(X, y)

# --- رسم درخت تصمیم ---
plt.figure(figsize=(12,7))
plot_tree(clf, feature_names=X.columns, class_names=clf.classes_, filled=True, rounded=True, fontsize=12)
plt.title(fa("🌳 درخت تصمیم برای تعیین پاداش کارکنان"))
plt.show()
```
