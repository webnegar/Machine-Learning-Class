```python
import numpy as np
import matplotlib.pyplot as plt

# داده‌ها
#x = np.array([1, 2, 3, 4, 5, 6 ,7])
#y = np.array([2, 4, 6, 9, 20 ,25, 30 ])

x = np.array([1, 2, 3, 4, 5])
y = np.array([30, 55, 60, 70, 100])

# پارامترهای اولیه
a = 0  # شیب اولیه
b = 0  # عرض از مبدا اولیه
alpha = 0.05  # نرخ یادگیری
iterations = 10  # تعداد تکرارها

# لیست برای ذخیره مقادیر تابع هزینه
losses = []
```

image.png

```python
# تابع هزینه
def compute_loss(x, y, a, b):
    n = len(x)
    y_predictions = a * x + b
    loss = (1/n) * np.sum(((y - y_predictions) ** 2))
    return loss
```

image.png

```python
# Gradient descent
for i in range(iterations):
    n = len(x)

    # Compute predictions
    y_predictions = a * x + b

    # Compute gradients
    grad_a = -(2/n) * np.sum(x * (y - y_predictions))
    grad_b = -(2/n) * np.sum(y - y_predictions)




    # Update a and b
    a = a - alpha * grad_a
    b = b - alpha * grad_b

    # Compute loss
    loss = compute_loss(x, y, a, b)
    losses.append(loss)

    # Print values every 100 iterations
    if i % 1 == 0:
      print ("grad_a:" , grad_a)
      print ("grad_b:" , grad_b)
      print(f"Iteration {i}: \t a = {a:.4f}, b = {b:.4f}, loss = {loss:.4f}")
      print ("-----------------------")
```

# New Section

```python
# Print final result
print(f"\nFinal result: a = {a:.4f}, b = {b:.4f}")

# Plot data and fitted line
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, a * x + b, color='red', label='Fitted line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Fitting a Line with Gradient Descent')
plt.show()

# Plot loss over iterations
plt.plot(losses)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Loss Over Iterations')
plt.show()
```
